```python
import apriltag
import cv2


input_vid = 'Video.mp4'
video = cv2.VideoCapture(input_vid)
status,img = video.read()
num = 0
status = True
h,w,l = img.shape
siz = (w,h)
op = cv2.VideoWriter('Output.avi',cv2.VideoWriter_fourcc(*'XVID'),30.0,siz)

while status:
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # options = apriltag.Detectoroptions(families='tag36h11',
    #                                  border=1,
    #                                  nthreads=4,
    #                                  quad_decimate=1.0,
    #                                  quad_blur=0.0,
    #                                  refine_edges=True,
    #                                  refine_decode=False,
    #                                  refine_pose=False,
    #                                  debug=False,
    #                                  quad_contours=True)

    detector = apriltag.Detector()
    result = detector.detect(gray)

    C = []
    for j in result:
        X = []
        Y = []
        for i in j.corners:
            x,y = i
            x = round(x)
            X.append(int(x))
            y = round(y)
            Y.append(int(y))
            img = cv2.circle(img,(int(x),int(y)), 3, (0,0,255), -1)
        cx = (X[0] + X[1] + X[2] + X[3])/4
        cy = (Y[0] + Y[1] + Y[2] + Y[3])/4
        #img = cv2.circle(img,(int(cx),int(cy)), 3, (0,0,255), -1)
        C.append((int(cx),int(cy)))
    
    for i in range(len(C)):
        if i+1 < len(C):
            cv2.line(img,C[i],C[i+1],(0,255,0),3)
        elif i+1 == len(C):
            cv2.line(img,C[i],C[0],(0,255,0),3)
        
    cv2.imshow("Output",img)
    if cv2.waitKey(5)&0xff == 27:
        break

    op.write(img)

    status,img = video.read()
    
cv2.destroyAllWindows()
video.release()
op.release()
```
